# axis = 1 ?
On souhaite supprimer une colonne, on indique donc l'indice 1 au paramètre axis. 1 correspond à l'axe des colonnes (0 correspond lui aux rangées).

```python
df_gender = df.copy()
```

Si on définit l'index sur le genre, on peut enlever toutes les rangées avec une certaine valeur en utilisant l'axe 0.

```python
df_gender.set_index("gender", inplace=True)
```

```python
df_gender.drop("Male", axis=0, inplace=True)
df_gender
```
